```python
import easyocr
from PIL import Image
```

```python
reader = easyocr.Reader(['bn', 'en'])
img = Image.open('data/1_english.jpg')
```

```python
result = reader.readtext(img)
print(result)
```

```python
for bbox, text, prob in result:
    print(f"Detected text: {text}, Confidence: {prob}, Bounding Box: {bbox}")
```
