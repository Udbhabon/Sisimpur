```python
import os
import google.generativeai as genai

# Set your API key (replace with your own from Google AI Studio)
os.environ["GOOGLE_API_KEY"] = "Give your Gemini API key from google ai studio"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
from PIL import Image
import io
```

```python
# Load your image
image_path = "data/1.jpg"
image = Image.open(image_path)

# Load the Gemini Pro Vision model
model = genai.GenerativeModel(model_name="models/gemini-1.5-flash")
```

```python
# Send image for OCR
response = model.generate_content([
    "Extract all the text from this image. The structure is:\n"
    "- Each question is followed by 4 answer options.\n"
    "- Maintain this structure and return the data in this JSON format:\n\n"
    "{\n"
    "  \"questions\": [\n"
    "    {\n"
    "      \"question\": \"...\",\n"
    "      \"options\": [\"A. ...\", \"B. ...\", \"C. ...\", \"D. ...\"]\n"
    "    },\n"
    "    {\n"
    "      \"question\": \"...\",\n"
    "      \"options\": [\"A. ...\", \"B. ...\", \"C. ...\", \"D. ...\"]\n"
    "    }\n"
    "  ]\n"
    "}",
    image
])


# Print extracted text
print(response.text)
```
