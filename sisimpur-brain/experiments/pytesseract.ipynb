```python
import pytesseract
from PIL import Image
import json
import re

# Set the path to tesseract executable if not in PATH
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Change for your system

# Load the image
image = Image.open('data/1.jpg')

# OCR with Bangla language
text = pytesseract.image_to_string(image, lang='ben')

# Post-process: Extract Q&A pattern
lines = [line.strip() for line in text.split('\n') if line.strip()]
questions = []
i = 0

while i < len(lines):
    question = lines[i]
    options = lines[i+1:i+5]
    if len(options) == 4:
        questions.append({
            "question": question,
            "options": options
        })
    i += 5

# Output as JSON
output = {"questions": questions}
with open('output.json', 'w', encoding='utf-8') as f:
    json.dump(output, f, ensure_ascii=False, indent=2)

print(json.dumps(output, ensure_ascii=False, indent=2))
```
